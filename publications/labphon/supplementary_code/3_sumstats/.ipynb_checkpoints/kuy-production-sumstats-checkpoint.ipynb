```python
import pandas as pd
import numpy as np
import itertools
import matplotlib.pyplot as plt
from matplotlib.transforms import offset_copy
import seaborn as sns

sns.set(font_scale = 1.5, style = 'whitegrid')
sns.plotting_context(rc={"grid.linewidth": 0.5})
```

```python
df = pd.read_csv('../0_data/kuy-production-clean.csv')

# Add generation
df['generation'] = df['age'].apply(lambda a: 'younger' if a < 45 else 'older')

# Merge with stimuli
stim = pd.read_csv('../0_data/kuy-production-stimuli.csv').fillna('')

coda_to_type = {'none':r'-$\emptyset$/-N', 
                'ŋ':r'-$\emptyset$/-N', 'm':r'-$\emptyset$/-N', 'n':r'-$\emptyset$/-N', 'l':r'-$\emptyset$/-N',
                'ʔ':'-ʔ',
                'h':'-h',
                'p':'T', 't':'T'}

# Group onsets and codas
stim['onset_type'] = stim['onset'].apply(lambda o: 'fricative' if o[0] == 's' else ('sonorant' if o[0] == 'l' else 'stop'))
stim['coda_type'] = stim['coda'].map(coda_to_type)
stim['Onset & Register'] = (stim['onset_type'] + ' (' + stim['aspiration'] + '), ' + stim['register']).str.replace(' ()', '', regex = False)
stim['Rhyme or Coda'] = stim.apply(lambda row: '-V:' + row['coda_type'] if ((row['vowelqual'][-1] == ':') & (row['coda_type'] == 'T'))
                                   else ('-V' + row['coda_type'] if row['coda_type'] == 'T' else row['coda_type']), axis = 1)

# Merge
stim_cols = stim[['word', 'vowelqual', 'Onset & Register', 'Rhyme or Coda']].rename(columns = {'vowelqual':'ipa'})

df = df.merge(stim_cols, on = 'word')
```

# Vowel durations
---
- [Contents](#Contents)
- [Vowel f0](#Vowel-f0)
- [f0 by onset and register](#f0-by-onset-and-register)
- [f0 by rhyme and register](#f0-by-rhyme-and-register)

```python
df.groupby(['Filename', 'ipa', 'register']).duration.mean().groupby(['ipa', 'register']).agg(['mean', 'std', 'count']).sort_values(by = 'ipa')
```

```python
plt.figure(figsize = (7, 7))

sns.violinplot(data = df.groupby(['Filename', 'ipa', 'register']).duration.mean().reset_index(),
               x = 'duration',
               y = 'ipa',
               order = ['i:', 'u:', 'e:', 'o:', 'ɛ:', 'a', 'ɑ'],
               hue = 'register',
               hue_order = ['modal', 'breathy'],
               split = True)

plt.xlabel('Duration (ms)')
plt.ylabel('Vowel quality')
plt.xlim(0, 800)
plt.legend(title = 'Register')

sns.despine(bottom = True)

plt.savefig('vdur.png', dpi = 300, bbox_inches = 'tight')
```

# Vowel f0
---
- [Contents](#Contents)
- [Vowel f0](#Vowel-f0)
- [f0 by onset and register](#f0-by-onset-and-register)
- [f0 by rhyme and register](#f0-by-rhyme-and-register)

```python
generations = ['younger', 'older']
vowels = ['i:', 'u:', 'e:', 'o:', 'ɛ:', 'a', 'ɑ']

fig, axes = plt.subplots(nrows = len(generations), ncols = len(vowels), figsize = (20, 7), tight_layout = True)

for i, (gen, v) in enumerate(itertools.product(generations, vowels)):

    ax = axes.flat[i]
    
    g = sns.lineplot(data = df.query('(generation == @gen) & (ipa == @v)'), x = 'time', 
                     y = 'f0_scaled', hue = 'register', hue_order = ['modal', 'breathy'], ax = ax,
                     legend = True)
    
    ax.set(xlim = (0, 20), yticks = np.arange(-0.6, 0.7, 0.2), xticks = np.arange(0, 20.0001, 5))
    
    if gen == 'younger': 
        ax.set(xlabel = None, title = v)
        ax.xaxis.set_ticklabels([])
        
    else:
        ax.set(xlabel = 'Time')

    if i not in [0, 7]:
        ax.set(ylabel = None)
        ax.yaxis.set_ticklabels([])
    
    else:
        ax.set(ylabel = gen + '\nScaled f0 values')

fig.supylabel('Generation', fontsize = 'x-large', x = 0.02)
fig.suptitle('Vowel', fontsize = 'x-large', y = 0.93)
        
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, title = 'Register', loc='lower center', ncol = 2, bbox_to_anchor=(0.5, -0.1))

[ax.get_legend().remove() for ax in axes.flat]
    
plt.savefig('vowelf0.png', dpi = 300, bbox_inches = 'tight')
```

# f0 by onset and register
- [Contents](#Contents)
- [Vowel durations](#Vowel-durations)
- [Vowel f0](#Vowel-f0)
- [f0 by rhyme and register](#f0-by-rhyme-and-register)

```python
g = sns.relplot(data = df, x = 'time', y = 'f0_scaled', hue = 'Onset & Register', 
            hue_order = ['stop (unaspirated), modal', 'sonorant, modal', 
                         'stop (aspirated), modal', 'sonorant, breathy', 'stop, breathy'
                         ],
            palette = [sns.color_palette('tab10')[i] for i in [0, 3, 2, 4, 1]],
            col = 'generation', kind = 'line', aspect = 1.5)

# Get handles and labels
handles = g._legend_data.values()
labels = g._legend_data.keys()
# Remove legend
g._legend.remove()
# Move it
g.fig.legend(handles=handles, labels = labels, frameon = False,
             loc='lower center', ncol=3, bbox_to_anchor=(0.42,-0.3), title = 'Onset & Register')

# Title
g.fig.suptitle('Generation', x = 0.4)

# Fill in axes
for i, gen in enumerate(['younger', 'older']):
    g.axes.flat[i].set(xticks = np.arange(0, 21, 5), xlim = (0, 20), xlabel = 'Time',
           yticks = np.arange(-1, 1.25, .25), ylim = (-1, 1),
           title = gen)
    
    if i == 0: g.axes.flat[i].set(ylabel = 'Scaled f0')

g.despine(right = False)
        
g.savefig('c-type-f0.png', dpi = 300, bbox_inches = 'tight')
```

# f0 by rhyme and register
- [Contents](#Contents)
- [Vowel durations](#Vowel-durations)
- [Vowel f0](#Vowel-f0)
- [f0 by onset and register](#f0-by-onset-and-register)

```python
generations = ['younger', 'older']
registers = ['modal', 'breathy']

fig, axes = plt.subplots(nrows = len(generations), ncols = len(registers), figsize = (15, 10), tight_layout = True)

for i, (gen, r) in enumerate(itertools.product(generations, registers)):

    ax = axes.flat[i]
    
    sns.lineplot(data = df.query('(generation == @gen) & (register == @r)'), x = 'time', 
                     y = 'f0_scaled', hue = 'Rhyme or Coda',
                     hue_order = [r'-$\emptyset$/-N', '-h', '-ʔ', '-VT', '-V:T'], 
                     palette = sns.color_palette('tab10')[0:5],
                     ax = ax, legend = True)
    
    ax.set(xlim = (0, 20), yticks = np.arange(-1, 1, 0.25), xticks = np.arange(0, 21, 5))
    
    if gen == 'younger': 
        ax.set(xlabel = None, title = r)
        ax.xaxis.set_ticklabels([])
        
    else:
        ax.set(xlabel = 'Time')

    if r == 'breathy':
        ax.set(ylabel = None)
        ax.yaxis.set_ticklabels([])
    
    else:
        ax.set(ylabel = gen + '\nScaled f0 values', yticks = np.arange(-1, 1.25, .25), ylim = (-1, 1))

# Large x and y labels
fig.supylabel('Generation', fontsize = 'x-large', x = 0.03)
fig.suptitle('Register', fontsize = 'x-large', x = 0.56, y = 0.95)

# Move labels to bottom
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, title = 'Rhyme or Coda', loc='lower center', ncol = 5, bbox_to_anchor=(0.56, -0.07))

[ax.get_legend().remove() for ax in axes.flat]

plt.savefig('vq-rhyme.png', dpi = 300, bbox_inches = 'tight')
```
